# Using the Terminal

You can also run Python files in the Terminal.

In VSCode (this editor), open a Terminal by pressing `` Ctrl+` ``, (the `` ` `` key is to the left of the `1` key).

You'll see a "prompt" that looks like this:

```
@your_github_username âžœ /workspaces/pbj-python-workshop (main) $ 
```

Let's break that down: 

- `@your_github_username` is the "user" using the terminal.
- `/workspaces/pbj-python-workshop` is the active path/directory that your terminal is in.
- `(main)` is the Git branch you're currently on - don't worry about this
- `$` denotes that the prompt is done and you can enter a command.

Commands are programs that run on your computer. You probably care about `python3`, which can run Python files!

## Try it

Give it a try! Type (**do not** copy/paste) this command into the Terminal: `python3 python/0_setup.py`.

`100` should print out.

This is often represented in documentation like:

```sh
$ python3 python/0_setup.py
100
```

Lines starting with `$` are commands (you type these), and the other lines are the expected output (what you should see).

## Some other commands

Some other commands you might care about include:

- `ls` - short for "list" - list files in your current directory
- `cd` - short for "change directory" - move directories. try `cd python` to enter the `python/` directory!
  - to go backwards, run `cd ..`
- `python3` - run a python file!

# FAQ: Terminal

## Why can't I copy/paste?

Believe it or not, the Control key used to send "control characters", which controlled the terminal. Control+C, for example, stops any running program.

You can right-click to copy/paste, though.

## Can I screw up stuff with this?

Depends on the command you enter. I haven't gone over any destructive commands here, though.
